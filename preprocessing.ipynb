```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
data = pd.read_csv('data/results.csv')

# Select matches only from the African Cup of Nations tournament
africa_cup_data = data[data['tournament'] == 'African Cup of Nations']

# Calculate goal difference and add it as a new column
africa_cup_data['goal_difference'] = africa_cup_data['home_score'] - africa_cup_data['away_score']

# Create the 'winner' column based on the goal difference
africa_cup_data['winner'] = africa_cup_data['goal_difference'].apply(lambda x: 1 if x > 0 else (2 if x < 0 else 0))

# Rearrange the columns
column_order = ['date', 'home_team', 'away_team', 'home_score', 'away_score', 'goal_difference', 'winner', 'tournament', 'city', 'country', 'neutral']
africa_cup_data = africa_cup_data[column_order]

# Data preprocessing steps (if necessary)
# ...

# Save the preprocessed data
africa_cup_data.to_csv('data/afcon.csv', index=False)

africa_cup_data.head(5)
```
